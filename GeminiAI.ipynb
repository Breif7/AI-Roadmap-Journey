```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# @title Numpy in action
temperatures_gdl = np.array([28,31,32,30,29,27,26])
print("Temperatures this week in Gudalajara")
print(temperatures_gdl)

average_temp = np.mean(temperatures_gdl)
print(f"The average temperature was: {average_temp:.2f}°C")
```

```python
# @title Pandas in action
sales_data = {
    'Product': ['Tequila', 'Birria', 'Torta Ahogada'],
    'Units Sold': [150,450,780]
}
df = pd.DataFrame(sales_data)
print("\n--- Sales Data ---")
print(df)
```

```python
# @title Matplotlib in action
plt.figure(figsize=(8,5))
plt.bar(df['Product'], df['Units Sold'], color=['#006400', '#CE1126', '#D2691E'])
plt.xlabel("Product")
plt.ylabel("Units Sold")
plt.title("Sales Performance of Guadalajara Staples")

plt.show()
```
